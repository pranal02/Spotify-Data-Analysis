```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks=pd.read_csv('tracks.csv')
df_tracks.head()
```

```python
# check for NULL Values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
# Top 10 least popular songs
sorted_df = df_tracks.sort_values('popularity',ascending=True).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
# top 10 songs based on popularity above 90
most_popular = df_tracks.query('popularity>90'  , inplace=  False).sort_values('popularity' , ascending=True)
most_popular[:10 ]
```

```python

# change the index of dataFrame
df_tracks.set_index("release_date" , inplace =True)
```

```python
#check artist present at 18 row
df_tracks[["artists"]].iloc[18]
```

```python
#here we delete the duration_ms column and created a new column as duration by app;ying lambda function to change to second
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x:round(x/1000))
df_tracks.drop("duration_ms" ,inplace =True , axis=1)
df_tracks.duration.head()
```

```python
corr_df = df_tracks.drop(["key","mode","explicit"],axis =1).corr(method="person")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.gegt_xticklabels(),rotation =90)
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df , y="loudness" , x = "energy" ,color="c").set(title="Loudness vs Energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df , y="popularity" , x = "acousticness" ,color="g").set(title="Popularity vs acousticness")
```

```python
df_genre=pd.read_csv("SpotifyFeatures.csv")
df_genre.head()
```

```python
plt.title("Duration of the Songs in Different Geners")
sns.color_palette("rocket",as_cmap = True)
sns.barplot(y='genre' , x ='duration_ms',data=df_genre)
plt.xlabel("Duration in milli Seconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous=df_genre.sort_values("popularity",ascending=False).head(10)
sns.barplot(y="genre" , x ="popularity" ,data=famous).set(title="Top 5 Genres by Popularity")

```
